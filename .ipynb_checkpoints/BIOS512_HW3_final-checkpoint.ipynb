Imagine we have a data frame called `data`, with a `type` column. Which one works and why?\
Function 1:
```

    df %>% group_by({{ column }}) %>% tally();
}

```

Function 2:
```
group_and_tally <- function(df, column){
    df %>% group_by(column) %>% tally();
}
group_and_tally(data, type);
```

Function 1 works since the column is embedded within the column and is then passed onto the function while Function 2 is not reading the read in column variable.

### Question 5

Download the patient_names.csv and patient_properties.csv files from Canvas and read them into R. Manually set the date columns to be date variables. Print the first 10 observations of each.

````r
patient_names <- read_csv("patient_names.csv")
patient_properties <- read_csv("patient_properties.csv")
````

````r
patient_names <- patient_names %>%
  mutate(BIRTHDATE = as.Date(BIRTHDATE, format = "%m/%d/%y"))
head(patient_names, 10)
````

````r
head(patient_properties, 10)
````

### Question 6
In the data frame pulled from patient_properties, you'll notice that the data is long, not wide. Do a pivot to make the properties their own columns. Print the first 10 observations after you do so.

````r
patient_properties_wide <- patient_properties %>%
  pivot_wider(
      id_cols = ID,
    names_from = property,  
    values_from = value  
  )

print(head(patient_properties_wide, 10))
````

### Question 7
Perform a left join of the names and properties_wide data frames by the ID column and print the first 10 rows.

````r
patients_joined <- patient_names %>% 
    left_join(patient_properties_wide, by = "ID")

print(head(patients_joined, 10))
````

Question 8
Notice something interesting about the names in our data set. Fix the name formatting and print the first 10 observations.

````r
head(patients_joined$FIRST, 20)
head(patients_joined$LAST, 20)
````

````r
library(stringr)

patient_names2 <- patients_joined %>%
  mutate(
    FIRST = str_remove(FIRST, "\\d+$"),
    LAST  = str_remove(LAST, "\\d+$")  
  )

print(head(patient_names2, 10))

head(patient_names2$FIRST, 20)
head(patient_names2$LAST, 20)
````

### Question 9
Using a for statement to loop through the categorical variables (excluding name and ID), print the counts of each unique value in descending order, using the mdpre() function for formatting.

````r
categorical_variables <- setdiff(names(patient_names2), c("FIRST", "LAST", "ID"))
    for (var in categorical_variables) { 
        cat("Counts for", var, ":\n") 
        counts <-sort(table(patient_names2[[var]]), decreasing = TRUE)
        mdpre(counts) 
        cat("\n")
    }
````

### Question 10
If you see any weird values, get rid of the ones that don't make sense, and combine the ones that are formatted wrong. Don't forget ot check the dates! Print the new tables for categorical values, and print the date ranges.

````r
cutoff <- as.Date("2020-01-01")

patient_names2$BIRTHDATE <- ifelse(
  patient_names2$BIRTHDATE > cutoff,
  patient_names2$BIRTHDATE %m-% years(100),
  patient_names2$BIRTHDATE
)

patient_names2$BIRTHDATE <- as.Date(patient_names2$BIRTHDATE)

summary(patient_names2$BIRTHDATE)
````

````r
patient_names3 <- patient_names2 %>%
  filter(MARITAL %in% c("M", "S")) %>%
  mutate(
    GENDER = recode(GENDER,
                    "Male" = "M",
                    "male" = "M",
                    "Female" = "F",
                    "female" = "F"),
    RACE = recode(RACE, 
                    "asiann" = "asian"),
    ETHNICITY = recode(ETHNICITY, 
                       "nonhispani" = "nonhispanic", 
                       "hispani" = "hispanic") 
 )
unique(patient_names3$GENDER)
unique(patient_names3$RACE)
unique(patient_names3$ETHNICITY)
unique(patient_names3$MARITAL)

patient_names3
````

Question 11
Make a histogram of the ages of patients by gender. 

````r
patient_names4 <- patient_names3 %>%
  mutate(AGE = as.numeric(interval(BIRTHDATE, Sys.Date()) / years(1)))

ggplot(patient_names4, aes(x = AGE, fill = GENDER)) + 
  geom_histogram(binwidth = 1, position = "dodge", color = "blue") + 
  labs(
    title = "Distribution of Patient Ages by Gender", 
    x = "Age in years", 
    y = "Number of Patients"
  )
````

Question 12
Make a scatterplot of birthdate by martial status.

````r
ggplot(patient_names4, aes(x = MARITAL, y = BIRTHDATE)) +
  geom_jitter(width = 0.2, height = 0, alpha = 0.6, color = "blue") +
  labs(title = "Scatterplot of Birthdate by Marital Status",
       x = "Marital Status",
       y = "Birthdate") +
  theme_minimal()
````
